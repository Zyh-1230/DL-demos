```python
import matplotlib.pyplot as plt
import numpy as np
import os # 文件系统操作
import PIL # 图像处理
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers, models

import pathlib # 路径操作
import tarfile # 解压文件
import urllib.request # 网络请求
```

```python
data_path = '/Users/zhouyuhan/Documents/Code/PythonForDataAnalysis/Test/DeepLearning EX100/flower_photos.tgz'
extract_path = '/Users/zhouyuhan/Documents/Code/PythonForDataAnalysis/Test/DeepLearning EX100/'

# 创建解压目录
os.makedirs(extract_path, exist_ok=True)

# 解压数据集
with tarfile.open(data_path, 'r:gz') as tar:
    tar.extractall(path=extract_path)
```

```python
# 设置数据目录
data_dir = pathlib.Path(extract_path) / 'flower_photos'

# 统计图像数量
image_count = len(list(data_dir.glob('*/*.jpg'))) # glob:模式匹配工具 
# 使用通配符匹配 *(1):匹配所有子目录 *.jpg:匹配所有以.jpg结尾的文件 glob()返回一个生成器 包含所有匹配文件的路径对象

print("图片总数为: ", image_count)
```

```python
# 查看玫瑰图片示例
roses = list(data_dir.glob('roses/*'))
PIL.Image.open(str(roses[0])) # 显示第一张玫瑰图片
```

```python
batch_size = 32
img_height = 180
img_width = 180
```

```python
# 创建训练集(80%)
train_ds = tf.keras.preprocessing.image_dataset_from_directory( # 从目录结构自动创建带标签的图像数据集
    data_dir,
    validation_split=0.2,
    subset='training',
    seed=123, # 确保可重现的分割
    image_size=(img_height, img_width),
    batch_size=batch_size
)
```

```python
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset='validation',
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size
)
```

```python
# 获取类别名称
class_names = train_ds.class_names # image_dataset_from_directory自动生成属性class_name
print(class_names)
```

```python
plt.figure(figsize=(20, 10))

for images, labels in train_ds.take(1):
    for i in range(20):
        ax = plt.subplot(5, 10, i+1)

        plt.imshow(images[i].numpy().astype('uint8'))
        plt.title(class_names[labels[i]])

        plt.axis('off')
```

```python
# 检查数据形状
for image_batch, labels_batch in train_ds:
    print(image_batch.shape)
    print(labels_batch.shape)
    break
```

```python
# 性能优化
AUTOTUNE = tf.data.AUTOTUNE # 特殊常量 -1 表示由TensorFlow运行时自动决定最优值
# cache(): 将数据集缓存到内存中 shuffle():打乱数据顺序 prefetch(): 异步数据预加载
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
num_classes = 5

model = models.Sequential([
    layers.Rescaling(1./255, input_shape=(img_height, img_width, 3)), # 标准化层: 将像素值缩放到0-1

    layers.Conv2D(16, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(64, (3, 3), activation='relu'),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes)
])

model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10)
```

```python
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'],  label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()

test_loss, test_acc = model.evaluate(val_ds, verbose=2)
```
